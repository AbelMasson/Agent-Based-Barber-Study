#### 1.2. DÃ©finition des Insectes (et de leur dÃ©placement)

##### 1.2.a. DÃ©finition des Insectes

On considÃ¨re une population d'insectes qui se dÃ©placent en marchant au sol. Les propriÃ©tÃ©s du dÃ©placement des insectes varient grandement d'une espÃ¨ce Ã  l'autre et sont modulÃ©es par de nombreux facteurs, comme par exemple la tempÃ©rature, la disponibilitÃ© des ressources trophiques [Humphries et al. 2010] ou la prÃ©sence d'autres insectes [Guy et al. 2008]. Dans notre cas, on suppose que le dÃ©placement d'un insecte Ã  chaque pas de temps est caractÃ©risÃ© uniquement par sa **direction** et la **distance** qu'il parcourt. Dans un souci de simplicitÃ©, on pose les hypothÃ¨ses suivantes : 

- On suppose que le **dÃ©placement** est **Ã©quiprobable dans toutes les directions**.
- On suppose que la **distance parcourue** est alÃ©atoire et tirÃ©e Ã  chaque pas de temps dans une loi normale. 
- On suppose enfin que **les insectes ne peuvent pas sortir du champ**. Lorsqu'ils atteignent le bord, ils "rebondissent" Ã  l'intÃ©rieur. 

Remarque : Les deux premiÃ¨res hypothÃ¨ses dÃ©finissent ce qu'on appelle une marche alÃ©atoire Ã  incrÃ©ment gaussien [https://www.math.ucla.edu/~caflisch/181.1.03f/Lect4-5.pdf]. 

Ainsi donc, en termes plus mathÃ©matiques, le dÃ©placement des insectes Ã  chaque pas de temps pourra Ãªtre dÃ©fini par un vecteur dont la direction est alÃ©atoire, et dont la norme (qui correspond Ã  la distance que va parcourir l'insecte) est tirÃ©e dans une loi normale.

<font color='green'> **Question 1** : La fonction *random_unit_vector()* ci-dessous renvoie un vecteur unitÃ© dont la direction est alÃ©atoire. En la modifiant lÃ©gÃ©rement, construisez une fonction *random_walk_vector()* qui renvoie un vecteur dont la direction est alÃ©atoire et la norme tirÃ©e dans une loi normale de moyenne mu et d'Ã©cart-type sigma. Les paramÃ¨tres mu et sigma reprÃ©senteront ici les capacitÃ©s de dispersion de l'insecte. </font>
    
<font color='green'> *Indice : la fonction np.random.normal(mu, sigma) renvoie une valeure tirÃ©e dans une loi normale de moyenne mu et d'Ã©cart-type sigma*  </font>

```python
def random_unit_vector() :
    
    """Gets a random point on the unit circle"""
    
    angle = random.random() * 2 * np.pi
    return np.cos(angle), np.sin(angle)
```

##### 1.2.b. Simulation du dÃ©placement

Maintenant que l'on a dÃ©fini les insectes et les rÃ¨gles regissant leur dÃ©placement, on peut simuler leur dÃ©placement sur plusieurs pas de temps. Dans cette section, on simule le dÃ©placement d'un insecte sur *num_steps = 100* pas de temps. Cette simulation nous donne la suite des localisations successives occupÃ©es par l'insecte sur le champ, que l'on stocke dans une matrice de taille *2 x num_steps*.

```python
num_steps = 100

def get_random_walk(num_steps, mu, sigma) :
    
    locations = np.zeros((2, num_steps))
    
    for i in range(1, num_steps):
        
        next_x, next_y = random_walk_vector(mu, sigma)
        locations[0, i] = abs(min(locations[0, i-1] + next_x, 100))
        locations[1, i] = abs(min(locations[1, i-1] + next_y, 100))
        
    return locations
```

#### 1.3. DÃ©finition des PiÃ¨ges (et de la capture)

On a dÃ©fini un champ, et des insectes qui se dÃ©placent dessus, plaÃ§ons maitenant des piÃ¨ges pour les attraper.

<font color='green'> **Question 2** : On suppose que l'on place un pot Barber au milieu du champ (sur le point de coordonnÃ©es [50,50]), et qu'il a un rayon de 0.1m. Ecrire une fonction *is_trapped(x,y)* qui prÃ©dit si l'insecte est capturÃ© ou non en fonction de sa localisation (x,y). </font>

<font color='green'> *Indice : Si l'insecte se situe dans le pot Barber, la distance entre sa localisation (x,y) et celle du pot (50, 50) est infÃ©rieure au rayon du pot.* </font>

```python
def is_trapped(location, location_Barber, radius_Barber) :
    
    x, y = location
    x_Barber, y_Barber = location_Barber
    
    return np.sqrt((x - x_Barber)**2 + (y - y_Barber)**2) <= radius_Barber
```

<font color='green'> **Question Subsidiaire** : On imagine maintenant que l'on place plusieurs pots Barber sur la parcelle. Ecrire la fonction *which_trap()* qui prÃ©dit si l'insecte est capturÃ© et renvoie le numÃ©ro du pot Barber dans lequel il est capturÃ© si oui et zÃ©ro sinon. </font>
    
<font color='green'> Tester la fonction pour n=3 Pots Barbers, placÃ©s sur un transect de parcelle aux coordonnÃ©es [10, 10], [10, 30] et [10, 50] numÃ©rotÃ©s 1, 2 et 3. </font>

```python
def which_trap(location, N_Barber, location_Barber, radius_Barber) :
    
    x, y = location
    x_Barber = [location_Barber[i][0] for i in range(N_Barber)]
    y_Barber = [location_Barber[i][1] for i in range(N_Barber)]
    
    liste_is_trapped = [(x - x_Barber[i])**2 + (y - y_Barber[i])**2 <= radius_Barber**2 for i in range(N_Barber)]
    
    if any(liste_is_trapped) :
        
        return liste_is_trapped.index(True) + 1 # (les pots sont numÃ©rotÃ©s de 1 Ã  3)
    
    return 0
```

### 2. Simulations

#### 2.1. DÃ©finition de la fonction de simulation 

On a un champ, des insectes et des piÃ¨ges... Tout ce qu'il nous faut ! Il s'agit maintenant de dÃ©finir une fonction de simulation c'est Ã  dire une fonction qui prend en entrÃ©e un nombre d'insectes, de piÃ¨ges (et leur positions), et un nombre de pas de temps, et qui prÃ©dit en sortie le nombre d'insectes piÃ©gÃ© dans chaque pot Barber.

<font color='green'> Pas de questions ici, parce que c'est un peu fastidieux... le code de la fonction de simulation est donnÃ©e ci-dessous </font>

```python
def Simulate_random_walk(num_steps, 
                         num_inds, 
                         mu, sigma, 
                         N_Barber, 
                         location_Barber, radius_Barber, 
                         shortcut_mesure=False) :
    
    # On initialise un dictionnaire pour stocker le dÃ©placement de chaque individu - Locationss - , et un dictionaire pour stocker 
    # l'instant t auquel l'individu i est piÃ©gÃ© (s'il l'est bien entendu) et dans quel Pot_Barber (tuple (t, pot)) - dict_Trapped. 
    
    Locationss = {} ; dict_Trapped = {}
    
    for individu in range(num_inds) :
        
        # On dÃ©finit au hasard une localisation initiale pour chaque individu
        locations = [[np.random.randint(0,100), np.random.randint(0,100)]]
        
        for i in range(1, num_steps):
        
            next_x, next_y = random_walk_vector(mu, sigma)
            new_location = [locations[i-1][0] + next_x, locations[i-1][1] + next_y]
            
            if bool(which_trap(new_location, N_Barber, location_Barber, radius_Barber)) :
                
                dict_Trapped[individu] = (i, which_trap(new_location, N_Barber, location_Barber, radius_Barber))
                locations[i:num_steps] = [[0,0]]*(num_steps - i + 1)
                
                break
                
            else :
            
                locations.append(new_location)
            
        Locationss[individu] = locations
    
    # Si on cherche Ã  simuler uniquement ce que l'on peut mesurer, i.e. le nombre d'individu piÃ©gÃ©s Ã  la fin de la simulation,
    # on peut s'arrÃªter rapidement ici en indiquant shortcut_mesure = True.
    
    if shortcut_mesure :
        return len(dict_Trapped)
    
    # Si des insectes on Ã©tÃ© piÃ©gÃ©s (i.e. dict_Trapped n'est pas vide), on construit une liste donnant le nombre d'individus 
    # dans le piÃ¨ge en fonction du temps, que l'on appelle Number_of_trapped_individuals_per_step.
        
    if bool(dict_Trapped) :
        
        Ntrapped_ind_per_step = []
        for i in range(1,N_Barber+1) :
            
            Ntrapped_ind_per_trap = [value[0] for value in dict_Trapped.values() if value[1] == i]
            Ntrapped_ind_per_step_per_trap = [sum([int(bool(k > value)) for value in Ntrapped_ind_per_trap])
                                                  for k in range(num_steps)]
            Ntrapped_ind_per_step.append(Ntrapped_ind_per_step_per_trap)
        
    return Locationss, Ntrapped_ind_per_step
```

On suppose que l'on a placÃ© ***N_Barber*** pots Barber aux points de coordonÃ©es ***location_Barber*** et qu'il ont tous le mÃªme rayon ***radius_Barber***.

*DÃ©finition de la fonction de rafraichissement de la figure Ã  chaque pas de temps*

```python
def update_plot(step_num):
    
    previous_locs = []
    for i in range(num_inds) :
        previous_locs = previous_locs + Locationss[i][max(0,step_num-5):step_num]
        
    current_locs = [Locationss[i][step_num] for i in range(num_inds)]
    
    previous_steps.set_data([previous_locs[i][0] for i in range(len(previous_locs))],
                            [previous_locs[i][1] for i in range(len(previous_locs))])
    current_steps.set_data([current_locs[i][0] for i in range(len(current_locs))],
                           [current_locs[i][1] for i in range(len(current_locs))])
    
    for i in range(N_Barber) :
        Hist_Pot_Barber[i].set_height(Number_of_trapped_individuals_per_step[i][step_num])
    
    timer_text1.set_text("time = {}".format(step_num))
    timer_text2.set_text("time = {}".format(step_num))
        
    return previous_steps, current_steps, Hist_Pot_Barber, timer_text1, timer_text2
```

#### 2.3. Simulation d'une base de donnÃ©es

Dans la suite de ce TD, nous allons nous recentrer sur la question que nous nous posions initialement Ã  savoir :  ***Est-il possible d'infÃ©rer le nombre d'individu prÃ©sent sur la parcelle, Ã  partir du nombre d'individus piÃ©gÃ©s et de leur capacitÃ© de dispersion ?***

A cette question nous allons rÃ©pondre en utilisant deux jeux de donnÃ©es simulÃ©s, un premier simulant des relevÃ©s de pots Barber (nombre d'individus piÃ©gÃ©s), et un second simulant des mesures de capacitÃ©s de dispersion. Afin d'enrichir lÃ©gÃ¨rement la question de recherche, nous allons simuler de surcroit trois habitats diffÃ©rents (prairie, champ cultivÃ©, ou un sol nu par exemple), sur lesquels le **nombre** et les **capacitÃ©s de dispersion** des insectes seront diffÃ©rents. La taille des bases de donnÃ©es simulÃ©es pourrra Ãªtre modifiÃ©e, et sera par dÃ©faut de 1000 parcelles (i.e. simulations) pour les donnÃ©es de piegeage, et de 200 mesures pour les donnÃ©es de dispersion.

***Protocole de simulation des donnÃ©es de piegeage***

On considÃ¨re 100 parcelles, sur lesquels on pose 3 pots Barber (sur le transect [[80, 10], [80, 30], [80, 50]]). 

On suppose que les **parcelles** sont **de trois types diffÃ©rents 1, 2, et 3** (e.g. une prairie, un champ cultivÃ©, et un sol nu) et que les capacitÃ©s de dispersion de l'insecte sont diffÃ©rentes sur chaque parcelle (par exemple, en prairie l'insecte se dÃ©place plus lentement que dans le champs, et dans le champs plus lentement que sur le sol nu). Ainsi pour chaque type de parcelle on dÃ©finit des **capacitÃ©s de dispersion diffÃ©rentes** (voir la fonction *random_walk_vector()* plus haut) :

- (mu_1, sigma_1) = (3, 3)
- (mu_2, sigma_2) = (6, 3)
- (mu_3, sigma_3) = (7, 3)

On suppose Ã©galement que selon leur type, **les parcelles n'acccueillent pas toutes le mÃªme nombre d'insectes** (par exemple, la densitÃ© d'insectes dans les prairies et les champs cultivÃ©s peut Ãªtre supÃ©rieure Ã  celle sur sol nu). On suppose Ã©galement que ce nombre varie alÃ©atoirement entre parcelles d'un mÃªme type. Ainsi pour la simulation, les nombres d'insectes dans chaque parcelle sont tirÃ©s, selon le type de parcelle, dans les lois binomiales suivantes :

- N1 ~ Binom(N1 = 200, p = 0.95)
- N2 ~ Binom(N2 = 160, p = 0.95)
- N3 ~ Binom(N3 = 60, p = 0.95)

La base de donnÃ©es prendra forme d'un tableau (ou dataframe), avec 1000 lignes (une par parcelle) renseignant 7 colonnes ; le type de la parcelle, le nombre d'individus simulÃ©s sur la parcelle, et le nombre d'individus piÃ©gÃ©s pour chaque espÃ¨ce (somme du nombre d'individus piÃ©gÃ©s dans chaque pot), et les paramÃ¨tres de la simulation.

```python
# On entre les valeurs des paramÃ¨tres dÃ©crites ci-dessus dans un dictionnaire.

dico_paramÃ¨tres = {'Type_1' : {'N' : 200, 'mu' : 3, 'sigma' : 3},
                   'Type_2' : {'N' : 160, 'mu' : 6, 'sigma' : 3},
                   'Type_3' : {'N' : 60, 'mu' : 7, 'sigma' : 3}}

# On dÃ©finit une fonction permettant de simuler un tableau de donnÃ©es de N_parcelles,
# avec les valeurs de paramÃ¨tre contenues dans le dictionnaire.

def Simulate_data(N_parcelles, dico_paramÃ¨tres) :
    
    #On intitialise le tableau qui va accueillir les donnÃ©es simulÃ©es.
    simulated_data = pd.DataFrame({}, columns = ['Type', 'N', 'N_PiÃ©gÃ©s', 'mu', 'sigma'])
    
    for i in range(N_parcelles) :
        
        # On choisit un type au hasard, et on le stocke
        Type  = "Type_{}".format(np.random.randint(3) + 1)
        row = [Type]

        # On simule le nombre d'individus sur la parcelle selon son type
        N = np.random.binomial(dico_paramÃ¨tres[Type]['N'], p = 0.95)
        row.append(N)
        
        # On simule le nombre d'individus piÃ©gÃ©s dans les pots grÃ¢ce Ã  la fonction dÃ©finie plus haut
        N_ind_piÃ©gÃ©s = Simulate_random_walk(num_steps = 100,
                                            num_inds = N, 
                                            mu = dico_paramÃ¨tres[Type]['mu'], 
                                            sigma = dico_paramÃ¨tres[Type]['sigma'], 
                                            N_Barber = 3,
                                            location_Barber = [[80, 10], [80, 30], [80, 50]], 
                                            radius_Barber = 5,
                                            shortcut_mesure = True)
        
        row.append(N_ind_piÃ©gÃ©s)
        
        # On rÃ©cupÃ¨re les valeurs des paramÃ¨tres de dispersion
        row.append(dico_paramÃ¨tres[Type]['mu']) ; row.append(dico_paramÃ¨tres[Type]['sigma'])
        
        #On stocke toute ces informations dans une ligne du tableau, et on recommence. 
        simulated_data.loc[len(simulated_data)] = row
    
    return simulated_data

# On fait tourner la fonction dÃ©finie ci-dessus sur le nombre choisi de parcelles
# et avec les valeurs de paramÃ¨tre contenues dans le dictionnaire

N_parcelles = 1000
data_simulated = Simulate_data(N_parcelles, dico_paramÃ¨tres)
```

***Simulation des donnÃ©es de Dispersion***

En plus de la simulation des captures dans les pots Barber, on simule une expÃ©rience permettant d'obtenir la vitesse des insectes Ã  chaque pas de temps.  Pour plus de simplicitÃ©, cette expÃ©rience sera simulÃ©e directement avec les lois de probabilitÃ©s utilisÃ©es dans le simulateur. On suppose que l'on dispose de 200 mesures de la vitesse des insectes dans les diffÃ©rents types de parcelles.

```python
def Simulate_data_dispersion(N_parcelles, dico_paramÃ¨tres) :
    
    #On intitialise le tableau qui va accueillir les donnÃ©es simulÃ©es.
    simulated_data_dispersion = pd.DataFrame({}, columns = ['Type', 'v'])
    
    for i in range(N_parcelles) :
        
        # On choisit un type au hasard, et on le stocke
        Type  = "Type_{}".format(np.random.randint(3) + 1)
        row = [Type]

        # On simule le nombre d'individus sur la parcelle selon son type
        v = max(0, np.random.normal(dico_paramÃ¨tres[Type]['mu'], dico_paramÃ¨tres[Type]['sigma']))
        row.append(v)
        
        #On stocke toute ces informations dans une ligne du tableau, et on recommence. 
        simulated_data_dispersion.loc[len(simulated_data_dispersion)] = row
    
    return simulated_data_dispersion

N_parcelles = 200
simulated_data_dispersion = Simulate_data_dispersion(N_parcelles, dico_paramÃ¨tres)
```

### 3. Inversion du problÃ¨me

Dans la section prÃ©cÃ©dente, on a Ã©tablit que la mesure du nombre d'individus capturÃ©s dans le pot Barber ne permet pas d'accÃ©der directement au nombre d'individus prÃ©sent sur la parcelle. En cause un biais sur la capture du Ã  la diffÃ©rence de mobilitÃ© des individus dans les diffÃ©rentes parcelles qui engendre une capture prÃ©ferentielle des individus les plus mobiles. Dans toute cette section, nous allons chercher Ã  construire un modÃ¨le permettant de corriger ce biais. Ce modÃ¨le aura donc pour objectif d'infÃ©rer le nombre d'individus prÃ©sent sur la parcelle Ã  partir des mesures du nombre d'individus piÃ©gÃ©s et d'une mesure de leur capacitÃ© de dispersion.

#### 3.1. Construction du modÃ¨le

Nous cherchons dans cette section Ã  construire un rÃ©seau bayÃ©sien reprÃ©sentant les variables et processus latents qui sous-tendent la capture des insectes dans les pots Barber, c'est-Ã -dire le nombre d'insectes sur la parcelle (on pourra Ã©galement parler de densitÃ©) et leur mobilitÃ©. A ce stade un grand nombre de modÃ¨les sont envisageables (avec ou sans hierarchisation en particulier). Dans un souci de temps, nous avons choisi ici le plus simple d'entre eux :

- On admet que le nombre d'insectes piÃ©gÃ©s dans le pot $N_{piÃ©gÃ©s}$ suit une loi normale de moyenne $mu = (a + b.log(mu_v)).N$ - avec $a = 0.11$ et $b = 0.10$, et de prÃ©cision $tau = 0.1$

```python
def Model_Dispersion_et_DensitÃ©(data,
                                data_dispersion) :
    
    parcelles_idx = np.asarray(data.index)
    parcelles_dispersion_idx = np.asarray(data_dispersion.index)
    
    types_idx, types = pd.factorize(data["Type"], sort=True)
    
    types_de_parcelle_idx = np.asarray([list(types).index(type_) for type_ in data["Type"]])
    types_de_parcelle_dispersion_idx = np.asarray([list(types).index(type_) for type_ in data_dispersion["Type"]])
    
    coords = {"parcelles" : parcelles_idx,
              "parcelles_dispersion" : parcelles_dispersion_idx,
              "types" : types}
    
    # DEFINITION DU MODELE
    # -------------------------------------------------------------------------------------------------------------------------------------------------

    with pm.Model(coords=coords) as model:
        
        # DONNEES
        
        N_piÃ©gÃ©s_obs = pm.ConstantData("N_piÃ©gÃ©s_obs", data["N_PiÃ©gÃ©s"], dims="parcelles")
        v_obs = pm.ConstantData("v_obs", data_dispersion["v"], dims="parcelles_dispersion")
        
        # MODELE
        
        mu_v = pm.Normal("mu_v", mu = 3, tau = 0.1, dims = "types")
        N = pm.Normal("N", mu = 100, tau = 0.01, dims = "types")
        
        mu_N_piÃ©gÃ©s = pm.Deterministic("mu_N_piÃ©gÃ©s", 0.11 + 0.10*np.log(mu_v), dims = "types")
        
        # Confrontation aux donnÃ©es observÃ©es
        v = pm.Normal("v", 
                       mu = mu_v[types_de_parcelle_dispersion_idx], 
                       tau = 1, 
                       observed = v_obs,
                       dims="parcelles_dispersion")
        
        N_piÃ©gÃ©s_pred = pm.Normal("N_piÃ©gÃ©s_pred", 
                                   mu = mu_N_piÃ©gÃ©s[types_de_parcelle_idx],
                                   tau = 0.1,
                                   observed=N_piÃ©gÃ©s_obs, 
                                   dims="parcelles")
    
    return model
```

```python
model = Model_Dispersion_et_DensitÃ©(data = data_simulated,
                                    data_dispersion = simulated_data_dispersion)

Graph = pm.model_to_graphviz(model)
Graph
```
